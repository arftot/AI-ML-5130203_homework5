PERCEPTRON MODEL

```python
import numpy as np
import pandas as pd

# Number of data points
n_samples = 1000

# Generating data for class 0
np.random.seed(42) 
class0_x = np.random.uniform(0, 3, n_samples // 2)
class0_y = np.random.uniform(0, 3, n_samples // 2)
class0 = pd.DataFrame({'feature 1': class0_x, 'feature 2': class0_y, 'class': 0})

# Generating data for class 1
class1_x = np.random.uniform(4, 7, n_samples // 2)
class1_y = np.random.uniform(4, 7, n_samples // 2)
class1 = pd.DataFrame({'feature 1': class1_x, 'feature 2': class1_y, 'class': 1})

# Combining the data
data = pd.concat([class0, class1])
data = data.sample(frac=1).reset_index(drop=True) # shuffling the data

# Add random noise (optional, for more realism)
noise = np.random.normal(0, 0.5, (n_samples, 2))
data[['feature 1', 'feature 2']] += noise


print(data.head())
print(data.shape)
data.to_csv("datasets/linearly_separable_data.csv", index=False)
```

MULTILAYER PERCEPTRON MODEL

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42) 

n_samples = 1000

# Creating data for the outer circle (class 0)
radius_inner = 2
angles_inner = np.random.uniform(0, 2 * np.pi, n_samples // 2)
x_inner = radius_inner * np.cos(angles_inner) + np.random.normal(0, 0.5, n_samples // 2)
y_inner = radius_inner * np.sin(angles_inner) + np.random.normal(0, 0.5, n_samples // 2)
class0 = pd.DataFrame({'feature1': x_inner, 'feature2': y_inner, 'class': 0})

# Creating data for the outer circle (class 1)
radius_outer = 5
angles_outer = np.random.uniform(0, 2 * np.pi, n_samples // 2)
x_outer = radius_outer * np.cos(angles_outer) + np.random.normal(0, 0.5, n_samples // 2)
y_outer = radius_outer * np.sin(angles_outer) + np.random.normal(0, 0.5, n_samples // 2)
class1 = pd.DataFrame({'feature1': x_outer, 'feature2': y_outer, 'class': 1})


# Combining the data
data = pd.concat([class0, class1])
data = data.sample(frac=1).reset_index(drop=True) # Shuffling the data

plt.figure(figsize=(6, 6))
plt.scatter(data['feature1'], data['feature2'], c=data['class'])
plt.title("Non-linearly separable data")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

data.to_csv("datasets/nonlinearly_separable_data.csv", index=False)

print(data.head())
print(data.shape)
```
